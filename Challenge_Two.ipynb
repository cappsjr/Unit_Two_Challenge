```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/Users/cappsjr/Downloads/happiness_report.csv')
print(df)
```

```python
plt.plot(df['Happiness Score'], color = 'green')
plt.plot(df['Economy (GDP per Capita)'], color = 'orange')
plt.show()
"""This graph shows a comparison of the happiness score and the GDP per capita of the countries included in the 2015 World Happiness Report.  The green line represents the happiness score, and the orange line represents the GDP per capita.  By comparing these two lines side-by-side one notices that there is a general correlation between the decline in happiness and the decline in GDP per capita."""
```

```python
plt.scatter(x = df['Happiness Rank'], y = df['Economy (GDP per Capita)'])
plt.xlabel('Happiness Rank')
plt.ylabel('GDP per Capita')
plt.title('Happiness Rank and GDP per Capita')
plt.show()
"""In this graph, the Happiness Ranks of the countries is being compared to their GDP per Capitas.  Here again, we see a general correlation between happiness and GDP per capita.  The better the country's happiness rank (1 being the best), the higher its GDP per capita tends to be.  The graph also suggests that the correlation is stronger from ranks 1 to 50 or 60 because the points are more tightly clustered together, whereas from ranks 60 and higher the points tend to cluster less and spread out more."""
```

```python
plt.scatter(x = df['Happiness Rank'], y = df['Health (Life Expectancy)'])
plt.xlabel('Happiness Rank')
plt.ylabel('Life Expectancy')
plt.title('Happiness Rank and Life Expectancy')
plt.show()
"""In this graph, the Happiness Rank is being compared with the Health (Life Expectancy) of the countries involved in the 2015 World Happiness Report.  It seems to show a similar correlation to that of the previous graph comparing the Happiness Rank and GDP per Capita.  The better the country's happiness rank (1 being the best), the higher the life expectancy.  Such a correlation may be expected because the countries with better ranks also have higher GDP per capita, which allows the citizenry the ability to afford better nutrition and healthcare, but we will examine any correlation between GDP per Capita and Life Expectancy in the next graph. """
```

```python
plt.figure(figsize = (5, 10))

plt.subplot(2, 1, 1)
plt.scatter(x = df['Health (Life Expectancy)'], y = df['Economy (GDP per Capita)'])
plt.xlabel('Life Expectancy')
plt.ylabel('GDP per Capita')
plt.title('Economy and Life Expectancy')

plt.subplot(2, 1, 2)
plt.scatter(x = df['Happiness Rank'], y = df['Freedom'])
plt.xlabel('Happiness')
plt.ylabel('Freedom')
plt.title('Freedom and Happiness')
plt.show()
"""Here are two graphs.  One shows the relationship between Life Expectancy and GDP per Capita, and the other shows the relationship between Freedom and Happiness.  The results are fairly predictable.  As GDP per capita of a country increases, so does the life expectancy of the citizenry, and as freedom increases, so does happiness."""
```
